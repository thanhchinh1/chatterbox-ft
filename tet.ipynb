```python
import json

# 1. Định nghĩa các chữ cái tiếng Việt có dấu (không bao gồm các chữ cái Latin cơ bản đã có sẵn)
vietnamese_chars = (
    "àáảãạăằắẳẵặâầấẩẫậđèéẻẽẹêềếểễệìíỉĩịòóỏõọôồốổỗộơờớởỡợùúủũụưừứửữựỳýỷỹỵ"
    "ÀÁẢÃẠĂẰẮẲẴẶÂẦẤẨẪẬĐÈÉẺẼẸÊỀẾỂỄỆÌÍỈĨỊÒÓỎÕỌÔỒỐỔỖỘƠỜỚỞỠỢÙÚỦŨỤƯỪỨỬỮỰỲÝỶỸỴ"
)

# 2. Đường dẫn tới file tokenizer gốc vừa tải về
tokenizer_path = "pretrained_models/tokenizer.json"

# 3. Đọc file JSON
with open(tokenizer_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# Xác định vị trí của vocab (tùy cấu trúc JSON của repo)
if "vocab" in data:
    vocab = data["vocab"]
elif "model" in data and "vocab" in data["model"]:
    vocab = data["model"]["vocab"]
else:
    vocab = data # Trường hợp file JSON chỉ chứa dict vocab thuần

# 4. Tìm ID lớn nhất hiện tại để đếm tiếp
max_id = max(vocab.values())

# 5. Cập nhật các ký tự tiếng Việt chưa có vào vocab
added_count = 0
for char in vietnamese_chars:
    if char not in vocab:
        max_id += 1
        vocab[char] = max_id
        added_count += 1

# 6. Ghi đè lại file JSON an toàn
with open(tokenizer_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print(f"Đã thêm thành công {added_count} ký tự tiếng Việt vào vocab!")
```
